# Task1

## Dataset Creation

````python
# for reproducing the same dataset, each time the notebook is run
random.seed(1)

# creating dataset
x = []
for i in range(2000):
  x.append(random.uniform(-100, 100))
x = np.asarray(x).reshape(2000, 1)

y = np.multiply(x, x).reshape(2000, 1)

print(x.shape)
print(y.shape)
````

> Observations from the above cell: 
1. Python has many modules that can be used to generate random numbers from uniform distribution one such module is ``` random ```.
2. It is very important to set the seed in such assignments, because there is an irresistable need to run the cells again and again to fix any bugs, and if the data changes between the runs, then it would be more difficult bug to fix. ```random.seed()```
3. For element wise multiplication of two matrices or two vectors ```np.multiply()``` can be used.
4. Converting the python list to numpy array is very beneficial, one such benefit can be observed in this part of code ```y = np.multiply(x, x)```. ```np.asarray()``` can be used to convert python list to numpy array.

## FNN Construction and Training

1. Construct a single hidden layer neural network. Decide the best hyper-parameters for it.

````python
# for reproducing the same results, each time the notebook is run
tf.random.set_seed(1)
initializer=tf.keras.initializers.GlorotUniform(seed=1)
````

> Observations from the above cell:
- Note: For perfect reproducability, training seed and weight matrix seed are set to 1.
- For the whole notebook all the seeds are set to 1.

````python
#FNN construction
model = tf.keras.Sequential()
model.add(tf.keras.Input(shape=(1,1)))
model.add(tf.keras.layers.Dense(16, activation=tf.nn.relu, kernel_initializer=initializer))
model.add(tf.keras.layers.Dense(1, activation='linear', kernel_initializer=initializer))
model.summary()

#hyper parameters
lr = 0.01
epochs = 500
batch_size = 32 #default
````

> Observations from the above cell:
1. ```model.summary()``` function can be used to get the summary of the model, along with the parameters of the model.
2. The hyperparameters which are decided are 
```
learning rate = 0.01
epochs = 500
batch_size = 32(default batch size in keras)
```
3. Activation function used after each layer is ReLU(rectified linear unit), and the activation used after the output layer is linear. As the problem is linear regression.

2. Train this neural network using no more than 5000 epochs on the above generated dataset. This will be your baseline performance.

````python
#additional parameters of the model
mse = tf.keras.losses.MeanSquaredError()
rmse = tf.keras.metrics.RootMeanSquaredError()
msle = tf.keras.losses.MeanSquaredLogarithmicError()
loss = msle
optimizer = tf.keras.optimizers.SGD(learning_rate=lr)

#training model
model.compile(loss=loss, optimizer=optimizer, metrics=[rmse, mse])
model.fit(x, y, shuffle=True, epochs=epochs)
````

## Model Construction

1. Decide the hyper-parameters like number of layers, number of neurons in each layer, batch size, learning rate, etc.

````python
layers = 8 #hidden layers
neurons_per_layer = [256, 256, 256, 256, 128, 128, 128, 128]
batch_size = 64
lr = 0.05
epochs = 50
````

> Observations from the above cell:
1. The activation used for each layer is ReLU (Rectified Linear Unit) and the activation of output lahyer is softmax with 10 neurons for 10-class classification.

2. Split the training data into train-validation sets. Appropriately choose a train-validation ratio.

````python
# the train to val ratio is 2:1
train_images, val_images, train_labels, val_labels = train_test_split(train_images, train_labels, test_size = 0.2, random_state=1, shuffle=True)
````

3. Train and tune the network using this base structure on train-validation data. This will be your baseline performance.

````python
# for reproducing the same results, each time the notebook is run
tf.random.set_seed(1)
initializerGU=tf.keras.initializers.GlorotUniform(seed=1)
````

````python
model1 = tf.keras.Sequential()
model1.add(tf.keras.Input(shape=input_shape))
model1.add(tf.keras.layers.Flatten())
model1.add(tf.keras.layers.Dense(256, activation='relu', kernel_initializer=initializerGU))
model1.add(tf.keras.layers.Dense(256, activation='relu', kernel_initializer=initializerGU))
model1.add(tf.keras.layers.Dense(256, activation='relu', kernel_initializer=initializerGU))
model1.add(tf.keras.layers.Dense(256, activation='relu', kernel_initializer=initializerGU))
model1.add(tf.keras.layers.Dense(128, activation='relu', kernel_initializer=initializerGU))
model1.add(tf.keras.layers.Dense(128, activation='relu', kernel_initializer=initializerGU))
model1.add(tf.keras.layers.Dense(128, activation='relu', kernel_initializer=initializerGU))
model1.add(tf.keras.layers.Dense(128, activation='relu', kernel_initializer=initializerGU))
model1.add(tf.keras.layers.Dense(10, activation='softmax', kernel_initializer=initializerGU))

#additional parameters of the model
cc = tf.keras.losses.CategoricalCrossentropy()
loss = cc
optimizer = tf.keras.optimizers.Adam(learning_rate=lr)

#training model
model1.compile(loss=loss, optimizer=optimizer, metrics=['acc'])
history1 = model1.fit(train_images, train_labels, validation_data=(val_images, val_labels), batch_size=batch_size, shuffle=True, epochs=epochs)
````

> Observations from the above cell:
1. No matter how long I train the neural network(many number of epochs), or how dense I train the neural network(many layers and units in each layer), or whatever loss function or metrics I use, the categorical loss for train dataset is not reducing below `2` and the accuracy for train dataset is not rising above `0.10`. Same is the case with validation dataset.
2. For classification problem, the output layer must be softmax with the number of classes as the number of units, 10.



4. Plot the train and validation accuracy and loss curves. Lets call them the performance plots.

````python
plt.title('Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epochs')
plt.plot(history1.history['acc'])
plt.plot(history1.history['val_acc'])
plt.legend(['train', 'validation'])
plt.show()

plt.title('Loss')
plt.ylabel('Categorical Crossentropy loss')
plt.xlabel('epochs')
plt.plot(history1.history['loss'])
plt.plot(history1.history['val_loss'])
plt.legend(['train', 'validation'])
plt.show()
````
